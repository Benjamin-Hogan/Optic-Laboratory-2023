# Neighorhood of Focus
## Code

```python
# Importing Modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Image Irradiance Data Manipulation

```python
# Importing 1st part of Lens relay images
imagefile_focus = '/Users/benjaminhogan/Library/Mobile Documents/com~apple~CloudDocs/School/Optics/Chapter 3/Hood of focus Images/Focus.jpg'
A_1 = plt.imread(imagefile_focus)

imagefile_above_focus = '/Users/benjaminhogan/Library/Mobile Documents/com~apple~CloudDocs/School/Optics/Chapter 3/Hood of focus Images/Above Focus.jpg'
A_2 = plt.imread(imagefile_above_focus);

imagefile_below_focus = '//Users/benjaminhogan/Library/Mobile Documents/com~apple~CloudDocs/School/Optics/Chapter 3/Hood of focus Images/Below Focus.jpg'
A_3 = plt.imread(imagefile_below_focus);
```

```python
# Performing Normalization of data and cutting the data

# Focused Lens
A_1 = A_1/255                          
A_1 = np.mean(A_1[:,:,0:2],2)
print(A_1)
col_1 = 800
cut_1 = (A_1[:,col_1]);           
softcut_1 = np.convolve(cut_1,np.ones(10),'valid')/10

#  Above Focus Lens
A_2 = A_2/255;                        
A_2 = np.mean(A_2[:,:,0:2],2)
print(A_2)
col_2 = 678
cut_2 = (A_2[:,col_2])                     
softcut_2 = np.convolve(cut_2,np.ones(10),'valid')/10

# Below Focused Lense
A_3 = A_3/255;                          
A_3 = np.mean(A_3[:,:,0:2],2)
print(A_3)
col_3 = 650
cut_3 = A_3[:,col_3];                     
softcut_3 = np.convolve(cut_3,np.ones(10),'valid')/10
```

```python
# X Data for graphing
x_1 = np.linspace(0, 100, num = 2100, endpoint = True, retstep = False, dtype = None)
x_2 = np.linspace(0, 100, num = 2091 , endpoint = True, retstep = False, dtype = None)
```

```python
plt.figure()
plt.grid()
plt.plot(x_1,cut_1, label='Focused Lens');                      
plt.plot(x_1,cut_2, label='Above Focus Lens');
plt.plot(x_1,cut_3, label='Below Lens'); 
plt.xlabel('Distance Along Cut (%)')
plt.ylabel('(Irradiance')
plt.title('Neighborhood of Focus Rough')
plt.legend()
plt.xlim(35,65)
plt.savefig('Neighborhood of Focus Rough')
plt.show()

plt.figure()
plt.grid()
plt.plot(x_2,softcut_1, label='Focused Lens')                      
plt.plot(x_2,softcut_2, label='Above Focus Lens');                     
plt.plot(x_2,softcut_3, label='Below Lens');
plt.xlabel('Distance Along Cut (%)')
plt.ylabel('Irradiance')
plt.title('Neighborhood of Focus Smoothed')
plt.xlim(35,65)
plt.legend()
plt.savefig('Neighborhood of Focus Smoothed')
plt.show()
```
